# Подробный отчет по разработке ИИ-ассистента «Иру»

Данный документ представляет собой пошаговое описание процесса проектирования и реализации Telegram-бота, интегрированного с современными нейросетевыми моделями.

---

## 1. Концепция и Техническое задание

**Идея:** Создать не просто скрипт-автоответчик, а интеллектуальную систему, способную эмулировать живое общение.

**Ключевые требования комиссии:**
- Использование LLM (Large Language Models).
- Работа с мультимедиа (голосовые сообщения).
- Демонстрация архитектурных навыков (чистый код, модульность).
- Безопасность и надежность данных.

## 2. Пошаговое описание проделанной работы

### Этап 1: Инициализация и настройка окружения
- **Создание бота:** Регистрация через BotFather и получение API-токена.
- **Безопасность:** Реализована система хранения секретов через `.env`. Это критически важный аспект разработки — токены никогда не попадают в открытый код.
- **Зависимости:** Подготовка файла `requirements.txt` для быстрой развертки проекта на любой машине.

### Этап 2: Проектирование архитектуры
Мы отказались от хранения всего кода в одном файле. Была выбрана **модульная архитектура**:
- **Config:** Централизованные настройки.
- **Services:** Отдельные классы для работы с ИИ и Голосом (инкапсуляция логики).
- **Handlers:** Обработчики команд, отделенные от бизнес-логики.
- **Utils:** Вспомогательные функции (логирование).

### Этап 3: Интеграция Нейросетевого ядра (Transformer-based)
- **Выбор модели:** Изначально использовался DeepSeek-V3, позже проект был успешно мигрирован на **Qwen2.5-72B-Instruct** для обеспечения стабильности и бесплатного доступа.
- **Реализация памяти:** Разработан алгоритм хранения контекста диалога. Мы передаем модели историю последних сообщений, что позволяет ей "помнить" тему разговора. Это реализует полноценное состояние (stateful) диалога.

### Этап 4: Разработка модуля обработки аудио
Это один из самых сложных этапов, включающий:
1. **Загрузку:** Получение OGG файла с серверов Telegram.
2. **Конвертацию:** Использование библиотеки `soundfile` для перевода OGG в WAV (частота 16кГц).
3. **Распознавание:** Отправка аудиопотока в Google Speech Engine через библиотеку `SpeechRecognition`.
4. **Связку:** Передача распознанного текста в ИИ-ядро.

### Этап 5: Интеллектуальный UX и Визуализация
- **Sentiment Analysis:** Мы научили бота анализировать настроение текста. Бот отправляет запрос к LLM для определения эмоции и ставит соответствующую реакцию (эмодзи) на сообщение пользователя.
- **Статусы:** Добавлена отправка действия `typing` (печатает), чтобы пользователь видел активность бота во время генерации ответа нейросетью.

### Этап 6: Логирование и Аналитика
- Реализована кастомная система записи чатов в `chat_history.log`.
- Создан архив голосовых сообщений `data/voice_history/`. Это позволяет в будущем использовать данные для дообучения моделей или анализа качества сервиса.

## 3. Технические характеристики (для тех. специалистов)

| Компонент | Технология |
| :--- | :--- |
| **Ядро ИИ** | Qwen2.5 (72 Billion parameters) |
| **Архитектура** | Transformer (Self-Attention) |
| **API Gateway** | Hugging Face Inference API |
| **STT Engine** | Google Speech-to-Text |
| **Library** | pyTelegramBotAPI (Asynchronous Polling) |
| **Storage** | File-based logging (Persistent Storage) |

## 4. Результаты и выводы

В результате работы был создан высокоуровневый ИИ-ассистент. Проект демонстрирует навыки работы с внешними API, обработки потоковых данных (аудио), проектирования архитектуры ПО и использования передовых достижений в области машинного обучения (Transformers).
