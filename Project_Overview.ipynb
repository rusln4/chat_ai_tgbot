# Подробный отчет по разработке ИИ-ассистента «Иру»

Данный документ представляет собой пошаговое описание процесса проектирования и реализации Telegram-бота, интегрированного с современными нейросетевыми моделями и отказоустойчивой архитектурой.

---

## 1. Концепция и Техническое задание

**Идея:** Создать не просто скрипт-автоответчик, а интеллектуальную и **надежную** систему, способную эмулировать живое общение в любых условиях.

**Ключевые требования:**
- Использование LLM (Large Language Models).
- Работа с мультимедиа (голосовые сообщения).
- **Отказоустойчивость:** Гарантия ответа даже при сбое основного провайдера.
- Демонстрация архитектурных навыков (чистый код, модульность).
- Безопасность и надежность данных.

## 2. Пошаговое описание проделанной работы

### Этап 1: Инициализация и настройка окружения
- **Создание бота:** Регистрация через BotFather и получение API-токена.
- **Безопасность:** Реализована система хранения секретов через `.env`. Это критически важный аспект разработки — токены никогда не попадают в открытый код.
- **Зависимости:** Подготовка файла `requirements.txt` для быстрой развертки проекта.

### Этап 2: Проектирование модульной архитектуры
Была выбрана **модульная архитектура** для чистоты и поддерживаемости кода:
- **Config:** Централизованные настройки и ключи API.
- **Services:** Отдельные классы для работы с ИИ и Голосом (инкапсуляция логики).
- **Handlers:** Обработчики команд Telegram, отделенные от бизнес-логики.
- **Utils:** Вспомогательные функции (логирование).

### Этап 3: Разработка модуля обработки аудио
Реализован полный цикл обработки голосовых сообщений:
1. **Загрузка:** Получение OGG файла с серверов Telegram.
2. **Конвертация:** Использование библиотеки `pydub` для перевода OGG в WAV.
3. **Распознавание:** Отправка аудиопотока в Google Speech Engine через библиотеку `SpeechRecognition`.
4. **Связка:** Передача распознанного текста в ИИ-ядро.

### Этап 4: Внедрение отказоустойчивой трехуровневой архитектуры ИИ
Это ключевая особенность проекта, обеспечивающая его надежность. Система последовательно опрашивает три эшелона моделей:
1. **Эшелон 1 (Основная модель):** `Qwen/Qwen2.5-72B-Instruct` через **Hugging Face API**. Это мощная модель для качественных ответов.
2. **Эшелон 2 (Резервная модель):** `google/gemma-7b-it:free` через **OpenRouter API**. Включается, если основная модель недоступна (например, из-за лимитов).
3. **Эшелон 3 (Супер-резерв):** `openrouter/auto` через **OpenRouter API**. Это автоматический маршрутизатор, который сам находит любую рабочую бесплатную модель, если предыдущие два эшелона отказали. Гарантирует ответ почти в любой ситуации.

### Этап 5: Улучшение UX и персонализация
- **Анализ настроения:** Бот пытается определить эмоцию в сообщении пользователя и ставит соответствующую реакцию (эмодзи).
- **Визуальная обратная связь:** Добавлен статус `Печатает... ✍️` для информирования пользователя об обработке запроса.
- **Персонализация:** Изменен системный промпт для придания боту неформального, дружеского стиля общения (обращения "Братан", "Дружище").

### Этап 6: Логирование и запуск
- Реализована система записи истории чатов и ошибок ИИ.
- Создан надежный скрипт `run.py`, который использует `pkill` для предотвращения запуска дубликатов бота.

## 3. Технические характеристики

| Компонент | Технология |
| :--- | :--- |
| **Ядро ИИ** | **Трехуровневая система:**<br>1. Qwen2.5-72B (HF)<br>2. Gemma 7B (OpenRouter)<br>3. Auto-Router (OpenRouter) |
| **Архитектура** | Transformer (Self-Attention) |
| **API Gateway** | Hugging Face Inference API, OpenRouter API |
| **STT Engine** | Google Speech-to-Text |
| **UX Элементы** | Chat Actions (typing), Message Reactions, Markdown Formatting |
| **Надежность** | Fallback-логика, Graceful Shutdown |

## 4. Результаты и выводы

В результате работы был создан не просто ИИ-ассистент, а **высоконадежная интеллектуальная система**. Проект демонстрирует навыки работы с несколькими внешними API, обработки аудио, проектирования отказоустойчивой архитектуры и использования передовых достижений в области машинного обучения. Внедрение трехуровневой системы с резервированием моделей гарантирует стабильную работу бота даже в условиях сбоя отдельных сервисов.
